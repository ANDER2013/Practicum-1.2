```python
# Contamos cuántas celdas vacías tiene cada columna
nulos = df.isnull().sum().sort_values(ascending=False)

# Solo mostramos las columnas que SÍ tienen nulos
nulos_existentes = nulos[nulos > 0]

print("COLUMNAS CON VALORES NULOS:")
print("-" * 40)
if len(nulos_existentes) == 0:
    print("  No hay valores nulos")
else:
    for col, cantidad in nulos_existentes.items():
        pct = cantidad / len(df) * 100
        print(f"  {col:<35} {cantidad:>6,} nulos ({pct:.1f}%)")

print(f"\nTotal celdas vacías en todo el dataset: {nulos.sum():,}")
```
